```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\tiara.santoso\Downloads\Data Exports\data_cluster.xlsx'

sheet_name = 'Test'
df = pd.read_excel(file_path, sheet_name=sheet_name)
# df = df.drop(columns=['Decision Tree Cluster'])
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler


df = df.drop(columns=['cluster'])
df.set_index('store_alias', inplace=True)

scaler = StandardScaler()
# df_scaled = df.drop(columns=['person_id'])
scaled_data = scaler.fit_transform(df)
```

```python
print(scaled_data)
```

```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns


n_clusters = 5
cluster = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
cluster.fit_predict(df)

# Visualization
# You can visualize the clusters by plotting a heatmap or using other visualization techniques
plt.figure(figsize=(10, 6))
sns.heatmap(df.T, cmap='viridis', linewidths=0.5, annot=True, fmt=".2f", cbar=False)
plt.title('Heatmap of Attributes')
plt.xlabel('Observations')
plt.ylabel('Attributes')
plt.show()

# Interpretation
# Analyze the clusters formed at different levels of the hierarchy
# You can examine the average values of each attribute within each cluster, visualize the clusters, or analyze any patterns that emerge

# For example, you can calculate the mean of each attribute for each cluster
cluster_means = df.groupby(cluster.labels_).mean()
print("Cluster Means:")
print(cluster_means)
```

```python
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the linkage matrix
linkage_matrix = sch.linkage(df, method='ward')

# Get the order of leaves in the dendrogram
dendrogram_leaves = sch.dendrogram(linkage_matrix, no_plot=True)['leaves']

# Check if dendrogram leaves are within bounds of the dataframe index
if max(dendrogram_leaves) >= len(df.index) or min(dendrogram_leaves) < 0:
    raise IndexError("Dendrogram leaves are out of bounds of the dataframe index")

# Reorder dataframe based on dendrogram leaves
df_reordered = df.iloc[dendrogram_leaves]

# Plot the dendrogram with colors for each cluster
plt.figure(figsize=(12, 6))
dendrogram = sch.dendrogram(linkage_matrix, labels=df.index[dendrogram_leaves], leaf_rotation=90, color_threshold=0.4*max(linkage_matrix[:, 2]))
plt.title('Dendrogram with Cluster Colors')
plt.xlabel('Observations')
plt.ylabel('Distance')
plt.show()

# Plot heatmap with reordered rows and columns
plt.figure(figsize=(12,12))
sns.heatmap(df_reordered, cmap='viridis', linewidths=0.5, annot=True, fmt=".2f", cbar=False)
plt.title('Heatmap of Attributes')
plt.xlabel('Observations')
plt.ylabel('Attributes')

plt.tight_layout()
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the linkage matrix
linkage_matrix = sch.linkage(df, method='ward')

# Create a figure with a 2x2 grid layout
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot the dendrogram with colors for each cluster
plt.subplot(2, 2, (1, 2))
dendrogram = sch.dendrogram(linkage_matrix, labels=df.index, leaf_rotation=90, color_threshold=0.4*max(linkage_matrix[:, 2]))
plt.title('Dendrogram with Cluster Colors')
plt.xlabel('Observations')
plt.ylabel('Distance')

# Reorder rows and columns based on dendrogram leaves
dendrogram_leaves = sch.dendrogram(linkage_matrix, no_plot=True)['leaves']
df_reordered = df.iloc[dendrogram_leaves]
df_reordered = df_reordered.iloc[dendrogram_leaves, :]

# Plot heatmap with reordered rows and columns
plt.subplot(2, 2, (3, 4))
sns.heatmap(df_reordered, cmap='viridis', linewidths=0.5, annot=True, fmt=".2f", cbar=False)
plt.title('Heatmap of Attributes with Cluster Ordering')
plt.xlabel('Observations (Clustered)')
plt.ylabel('Attributes (Clustered)')

# Plot original heatmap for comparison
plt.subplot(2, 2, 3)
sns.heatmap(df, cmap='viridis', linewidths=0.5, annot=True, fmt=".2f", cbar=False)
plt.title('Original Heatmap of Attributes')
plt.xlabel('Observations')
plt.ylabel('Attributes')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Define the features and their weights
features = ['price_cluster', 'aov_cluster', 'order / sqm - quartile', 'brand_cluster', 'net rev / sqm - quartile', 'category_cluster']
weights = [3, 4, 2, 2, 2, 1]  # Weights corresponding to each feature

# Apply the weights to the features
for i, feature in enumerate(features):
    df[feature] = df[feature] * weights[i]

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Perform k-means clustering
k = 5  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Print cluster centers
print("Cluster centers:")
print(scaler.inverse_transform(kmeans.cluster_centers_ / weights))

# Optionally, you can also save the cluster labels back to your DataFrame
# df['cluster'] = kmeans.labels_
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# Apply t-SNE for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(df_scaled)

# Plot t-SNE visualization
plt.figure(figsize=(10, 8))
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue=df['cluster'], palette='Set1', legend='full')
plt.title('t-SNE Visualization of Clustering Results')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(scaled_data)

df['Cluster'] = kmeans.labels_
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
for cluster_label in range(5): 
    cluster_points = scaled_data[df['Cluster'] == cluster_label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_label}', alpha=0.5)

    # Compute centroid of the cluster
    centroid = np.mean(cluster_points, axis=0)
    
    # Connect data points to centroid
    for point in cluster_points:
        plt.plot([centroid[0], point[0]], [centroid[1], point[1]], linestyle='-', color='gray', alpha=0.3)

    # Plot centroid
    plt.plot(centroid[0], centroid[1], marker='o', markersize=8, color='black')

plt.title('KMeans Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# Assuming `scaled_data` and `df['Cluster']` are defined

# Reduce the dimensionality of the data to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_data = tsne.fit_transform(scaled_data)

# Plotting clusters using t-SNE
plt.figure(figsize=(8, 6))
for cluster_label in range(5): 
    # Change cluster label for legend
    new_cluster_label = cluster_label + 1
    
    cluster_points = tsne_data[df['Cluster'] == cluster_label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {new_cluster_label}', alpha=0.5)

    # Compute centroid of the cluster
    centroid = np.mean(cluster_points, axis=0)
    
    # Connect data points to centroid
    for point in cluster_points:
        plt.plot([centroid[0], point[0]], [centroid[1], point[1]], linestyle='-', color='gray', alpha=0.3)

    # Plot centroid
    plt.plot(centroid[0], centroid[1], marker='o', markersize=8, color='black')

plt.title('Decision Tree Clustering with t-SNE')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend()
plt.grid(True)
plt.show()
```
