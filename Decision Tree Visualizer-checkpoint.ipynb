```python
import pandas as pd

# Extract decision rules
tree_rules = export_text(tree_clf, feature_names=list(data.columns)).split('\n')

# Initialize a list to store decision rules
decision_rules = []

# Parse decision rules and store them in a list
for rule in tree_rules:
    if '|---' in rule:
        rule = rule.replace('|--- ', '')
        parts = rule.split()
        if len(parts) >= 4:
            feature, condition, value, outcome = parts[0], parts[1], parts[2], ' '.join(parts[3:])
            decision_rules.append([feature, condition, value, outcome])
        else:
            decision_rules.append([rule, None, None, None])

# Create a Pandas DataFrame to display decision rules
decision_rules_df = pd.DataFrame(decision_rules, columns=['Feature', 'Condition', 'Value', 'Outcome'])

# Display the decision rules
print(decision_rules_df)
```

```python
import numpy as np
import plotly.graph_objects as go
from sklearn.tree import DecisionTreeClassifier, export_text

# Assuming `data` contains your feature matrix and `cluster_labels` contains the cluster assignments
data = df[['aov_cluster','order_sqm_quartile','price_cluster','netrev_sqm_quartile','brand_cluster']]
cluster_labels = df['cluster']  # Adjust this according to your DataFrame

# Fit a decision tree classifier
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(data, cluster_labels)

# Export decision rules
decision_rules = export_text(tree_clf, feature_names=list(data.columns))

# Function to parse decision rules and create Plotly-compatible tree structure
def parse_decision_rules(decision_rules):
    lines = decision_rules.split('\n')
    tree_structure = {'feature': [], 'value': [], 'outcome': []}
    for line in lines:
        if line.startswith('|---'):
            parts = line.split()
            tree_structure['feature'].append(parts[0])
            tree_structure['value'].append(float(parts[2]))
            tree_structure['outcome'].append(parts[-1])
    return tree_structure

# Parse decision rules
tree_structure = parse_decision_rules(decision_rules)

# Function to create Plotly decision tree figure
def create_plotly_decision_tree(tree_structure):
    fig = go.Figure()
    for i in range(len(tree_structure['feature'])):
        fig.add_trace(go.Scatter(x=[tree_structure['feature'][i]], y=[i+1],
                                  mode='markers+text',
                                  marker=dict(color='blue', size=10),
                                  text=[f"Value: {tree_structure['value'][i]}<br>Outcome: {tree_structure['outcome'][i]}"],
                                  textposition="bottom center"))
        if i != 0:
            fig.add_trace(go.Scatter(x=[tree_structure['feature'][i-1], tree_structure['feature'][i]],
                                      y=[i, i+1],
                                      mode='lines',
                                      line=dict(color='blue', width=2)))
    fig.update_layout(title='Decision Tree Visualization',
                      xaxis=dict(title='Feature'),
                      yaxis=dict(title='Depth', autorange='reversed'),
                      showlegend=False)
    return fig

# Create Plotly decision tree figure
decision_tree_fig = create_plotly_decision_tree(tree_structure)

# Show decision tree figure
decision_tree_fig.show()
```
