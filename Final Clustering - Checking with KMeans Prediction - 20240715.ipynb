```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

file_path = r"C:\Users\tiara.santoso\Downloads\Data Exports\raw_data_final_segmentation_20240715.xlsx"
data = pd.read_excel(file_path, sheet_name='test')
data.set_index('store_alias', inplace=True)

data.head(20)
```

```python
features = ['price_point', 'aov&volume_point', 'order/sqm quartile', 'net_rev_quartile']
X = data[features]

Z = linkage(X, method='complete')
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Store Alias')
plt.ylabel('Distance')
dendrogram(Z, labels=data.index.tolist(), leaf_rotation=90, leaf_font_size=10)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

max_d = 3.5
clusters = fcluster(Z, max_d, criterion='distance')
data['cluster'] = clusters

print(data[['cluster']])

```

```python
output_file_path = r"C:\Users\tiara.santoso\Downloads\Data Exports\store_clusters.xlsx"
data.to_excel(output_file_path, sheet_name='ClusteredData')

print(data[['cluster']])
```
