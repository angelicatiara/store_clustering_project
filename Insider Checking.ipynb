```python


# Assuming df is your DataFrame

# Create a DataFrame to store the summary counts
event_name_summary = pd.DataFrame({
    'Event Name': df['event_name'].unique()  # Get unique event names
})

# Count total occurrences for each event_name
total_counts = df['event_name'].value_counts().reset_index()
total_counts.columns = ['Event Name', 'Total Count']

# Count non-null (actual) values
non_null_counts = df['event_name'].notna().groupby(df['event_name']).sum().reset_index()
non_null_counts.columns = ['Event Name', 'Non-null (Actual) Count']

# Count NaN values
nan_counts = df['event_name'].isna().groupby(df['event_name']).sum().reset_index()
nan_counts.columns = ['Event Name', 'Null/NaN Count']

# Merge the counts into the summary DataFrame
event_name_summary = event_name_summary.merge(total_counts, on='Event Name', how='left')
event_name_summary = event_name_summary.merge(non_null_counts, on='Event Name', how='left')
event_name_summary = event_name_summary.merge(nan_counts, on='Event Name', how='left')

# Fill NaN with 0 for counts
event_name_summary.fillna(0, inplace=True)

# Convert counts to integers
event_name_summary['Total Count'] = event_name_summary['Total Count'].astype(int)
event_name_summary['Non-null (Actual) Count'] = event_name_summary['Non-null (Actual) Count'].astype(int)
event_name_summary['Null/NaN Count'] = event_name_summary['Null/NaN Count'].astype(int)

# Display the result
print(event_name_summary)
```
