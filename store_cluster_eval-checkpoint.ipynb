Initial Dataset Exploration & Variables Correlation using Pandas Profiling 

```python
# --- Dataset Report ---
ProfileReport(df, title='Store Dataset Report', minimal=True, progress_bar=False, samples=None, correlations=None, interactions=None, explorative=True, dark_mode=True, notebook={'iframe':{'height': '600px'}}, html={'style':{'primary_color': '#D0EE11'}}, missing_diagrams={'heatmap': False, 'dendrogram': False}).to_notebook_iframe()
```

```python
# --- Corr. Map Variables ---
df_corr = df.drop('store_name', axis=1).copy()  # Exclude 'store_name' column

# One-hot encode categorical variables
df_corr = pd.get_dummies(df_corr, columns=['area_category', 'city_type'], drop_first=True)

# --- Correlation Map ---
mask = np.triu(np.ones_like(df_corr.corr(), dtype=bool))
fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(df_corr.corr(), mask=mask, annot=True, cmap='GnBu', linewidths=0.2, cbar=False, annot_kws={"size": 6})
yticks, ylabels = plt.yticks(rotation=0)
xticks, xlabels = plt.xticks(rotation=45)
ax.set_xticklabels(xlabels, fontsize=6)
ax.set_yticklabels(ylabels, fontsize=6)
plt.suptitle('Correlation Map of Variables', fontweight='heavy', x=0.18, y=0.965, ha='left', fontsize=12)
plt.title('Insert Title Here\n', fontsize=7, loc='left', style='italic')
plt.tight_layout(rect=[0, 0.04, 1, 1.01])
plt.gcf().text(0.84, 0.03, 'kaggle.com/caesarmario', style='italic', fontsize=5);
plt.show();
```
